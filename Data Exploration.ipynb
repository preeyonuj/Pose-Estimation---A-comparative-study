# Smurf-tastic

```python
cnt = 0
class_index = "A026"

for file in glob.glob(path+"*"):
    if class_index in file:
        cnt += 1
        img = cv2.imread(file, )
        
        plt.imshow(img)
        plt.show()
        
        if cnt == 20:
            break
```
